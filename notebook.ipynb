```python
import h5py

# Mở file HDF5
with h5py.File('./Baseline/PURE.hdf5', 'r') as file:
    
    # Hàm này sẽ được gọi cho mỗi nhóm/dataset
    def print_details(name, obj):
        print(f"Name: {name}")
        # Kiểm tra nếu đối tượng là một dataset
        if isinstance(obj, h5py.Dataset):
            print(f"Dataset details:")
            print(f" - Shape: {obj.shape}")  # In kích thước của dataset
            print(f" - Dtype: {obj.dtype}")  # In kiểu dữ liệu của dataset
            print(f" - Data: {obj[...]}")    # In dữ liệu của dataset (cẩn thận với dataset lớn)
        elif isinstance(obj, h5py.Group):
            print(f"Group")
        print()

    # Duyệt qua tất cả các nhóm và datasets, áp dụng hàm print_details cho mỗi đối tượng
    file.visititems(print_details)
```
