```python
import h5py

# M file HDF5
with h5py.File('./Baseline/PURE.hdf5', 'r') as file:
    
    # Hm ny s c gi cho mi nhm/dataset
    def print_details(name, obj):
        print(f"Name: {name}")
        # Kim tra nu i tng l mt dataset
        if isinstance(obj, h5py.Dataset):
            print(f"Dataset details:")
            print(f" - Shape: {obj.shape}")  # In kch thc ca dataset
            print(f" - Dtype: {obj.dtype}")  # In kiu d liu ca dataset
            print(f" - Data: {obj[...]}")    # In d liu ca dataset (cn thn vi dataset ln)
        elif isinstance(obj, h5py.Group):
            print(f"Group")
        print()

    # Duyt qua tt c cc nhm v datasets, p dng hm print_details cho mi i tng
    file.visititems(print_details)
```
